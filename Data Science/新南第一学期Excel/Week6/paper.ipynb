```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create the data
data = {
    "Paper Size": ["A5", "A4", "A3", "A2", "A1", "A0"],
    "Boxes Ordered": [300, 600, 290, 200, 100, 250],
    "Cost per Box ($)": [1.50, 2.55, 3.20, 4.90, 6.00, 9.90]
}

df = pd.DataFrame(data)
df
```

```python
# Add total cost for each row
df['Total Cost ($)'] = df['Boxes Ordered'] * df['Cost per Box ($)']
df
```

```python
# Simple visualization of the data
plt.figure(figsize=(8, 5))
plt.bar(df['Paper Size'], df['Boxes Ordered'], color='skyblue')
plt.xlabel('Paper Size')
plt.ylabel('Boxes Ordered')
plt.title('Boxes Ordered by Paper Size')
plt.tight_layout()
plt.show()
```

```python
# Simple column chart: paper size vs total cost
plt.figure(figsize=(8, 5))
plt.bar(df['Paper Size'], df['Total Cost ($)'], color='lightgreen')
plt.xlabel('Paper Size')
plt.ylabel('Total Cost ($)')
plt.title('Total Cost by Paper Size')
plt.tight_layout()
plt.show()
```
