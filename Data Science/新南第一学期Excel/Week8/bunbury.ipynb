```python
import pandas as pd

data = {
    "Item": [
        "Phone",
        "Office Rental",
        "Leasing charges",
        "Petrol",
        "Maintenance",
        "Staff Salaries",
        "Drivers' Wages"
    ],
    "Actual": [2500, 2000, 3000, 7000, 4500, 8000, 16500],
    "Budget": [1200, 2000, 3000, 5000, 7000, 8000, 14000]
}

df = pd.DataFrame(data)
df
```

```python
df["Difference"] = df["Actual"] - df["Budget"]
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(len(df["Item"]))

plt.figure()

plt.bar(x - 0.2, df["Actual"], width=0.4, label="Actual")
plt.bar(x + 0.2, df["Budget"], width=0.4, label="Budget")

plt.xticks(x, df["Item"], rotation=90)
plt.ylabel("Amount")
plt.title("Actual vs Budget by Item")
plt.legend()

plt.show()
```
