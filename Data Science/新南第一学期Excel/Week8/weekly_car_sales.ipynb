```python
import pandas as pd

data = {
    "STATE": ["NSW", "NSW", "QLD", "QLD", "NSW", "NSW"],
    "REGION": ["Kingsford", "Atarmon", "Chermside", "Taringa", "Kensington", "Zetland"],
    "BMW": [2, 6, 9, 7, 16, 12],
    "PEUGEOT": [8, 5, 9, 22, 2, 0],
    "RENAULT": [5, 7, 9, 16, 11, 24],
    "SUBARU": [21, 23, 11, 23, 3, 9],
    "RANGE ROVER": [2, 3, 3, 15, 19, 24],
    "REGION TOTAL": [38, 44, 41, 83, 51, 69]
}

df = pd.DataFrame(data)

df
```

```python
state_df = df.groupby("STATE").sum(numeric_only=True)

state_df
```

```python
import matplotlib.pyplot as plt

state_df.plot(kind="bar")

plt.title("Car Sales by State")
plt.xlabel("State")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

```python
# Compare total car sales by brand, ignoring regions
car_columns = ["BMW", "PEUGEOT", "RENAULT", "SUBARU", "RANGE ROVER"]
sales_by_car = df[car_columns].sum()
sales_by_car
```

```python
sales_by_car.plot(kind="bar", color=["#4C72B0", "#55A868", "#C44E52", "#8172B2", "#CCB974"])

plt.title("Total Car Sales by Brand")
plt.xlabel("Car Brand")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Plot total car sales by region
region_df = df.groupby("REGION").sum()
region_df.plot(kind="bar", stacked=False, figsize=(10, 6))

plt.title("Car Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Plot total car sales by region only, ignoring individual car brands
region_totals = df.set_index('REGION')['REGION TOTAL']
region_totals.plot(kind='bar', color='skyblue', figsize=(10, 6))

plt.title('Total Car Sales by Region')
plt.xlabel('Region')
plt.ylabel('Region Total Sales')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Boxplot for each car brand, ignoring states and regions
df[['BMW', 'PEUGEOT', 'RENAULT', 'SUBARU', 'RANGE ROVER']].boxplot(
    figsize=(10, 6),
    patch_artist=True,  # 必须加这个才能上色
    boxprops=dict(facecolor='lightblue')  # 盒子颜色
)

plt.title('Boxplot of Car Sales by Brand')
plt.xlabel('Car Brand')
plt.ylabel('Sales')
plt.xticks(rotation=0, ha='right')
plt.tight_layout()
plt.show()
```
